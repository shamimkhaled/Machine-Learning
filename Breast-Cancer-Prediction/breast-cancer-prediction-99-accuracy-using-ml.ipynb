```python
# split the dataset into dependent(X) and Independent(Y) datasets
#x = df.drop(['id', 'diagnosis'], axis = 1).values
#y = df['diagnosis'].values
x=df.iloc[:,2:31].values
y=df.iloc[:,1].values
```

```python
y = pd.DataFrame(y)
print(y.value_counts())
#y.value_counts()
```

```python
#over_sampling
from imblearn.over_sampling import RandomOverSampler
over_sample = RandomOverSampler(random_state=100)
x, y =over_sample.fit_resample(x,y)
```

```python
y.value_counts()
```

```python
# spliting the data into trainning and test dateset
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import train_test_split, cross_val_score
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
x_train=StandardScaler().fit_transform(x_train)
x_test=StandardScaler().fit_transform(x_test)
```

# Build the Classification ML Model - GradientBoosting, Random Forest, Decision Tree, XGBoost, LightGBM,Logistic Regression, MLP

```python
#classify function
def classify(model):
    model.fit(x_train, y_train)
    #print("The accuracy of this model is: ", model.score(xtest, ytest)*100)
    predict = model.predict(x_test)
    print(classification_report(y_test, predict))
```

#### Gradient Boosting

```python
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier()
classify(clf)
```

```python
print('Gradient Boosting Classifier Confusion Matrix')
y_pred = clf.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True, fmt = 'g')
```

#### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rfcl = RandomForestClassifier(n_estimators=200, criterion='gini', min_samples_split=5, min_samples_leaf=2, max_features='auto', bootstrap=True, n_jobs=-1, random_state=42)
rfcl.fit(x_train, y_train)
```

```python
predict_rfcl = rfcl.predict(x_test)
print(classification_report(y_test, predict_rfcl))
```

```python
print('Random Forest Classifier Confusion Matrix')
y_pred = rfcl.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True, fmt = 'g')
```

 #### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_train, y_train)
predict_dt = dt.predict(x_test)
print(classification_report(y_test, predict_dt))
```

```python
print('Decision Tree Classifier Confusion Matrix')
y_pred = dt.predict(x_test)
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True)
```

#### XGBoost

```python
from xgboost import XGBClassifier
xgb = XGBClassifier()
xgb.fit(x_train, y_train)
```

```python
predict_xgb = xgb.predict(x_test)
print(classification_report(y_test, predict_xgb))
```

#### LightGBM

```python
from lightgbm import LGBMClassifier
lgbm = LGBMClassifier()
lgbm.fit(x_train, y_train)
```

```python
predict_lgbm = lgbm.predict(x_test)
print(classification_report(y_test, predict_lgbm))
```

```python
print('LGBM Classifier Confusion Matrix')
y_pred = lgbm.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, fmt = 'g')
```

### LightGBM ROC Curve

```python
from sklearn.metrics import roc_auc_score, roc_curve, auc
```

```python
y_preds = lgbm.predict(x_test).ravel()

lgbm_fpr, lgbm_tpr, thresholds = roc_curve(y_test, y_preds)
plt.figure(1)
plt.plot([0, 1], [0, 1], 'y--')
plt.plot(lgbm_fpr, lgbm_tpr, marker='.')
plt.xlabel('False positive rate')
plt.ylabel('True positive rate')
plt.title('Receiver Operating Charcateristic - ROC curve for LightGBM')
plt.show()
```

### AUC - Area Under Curve for all Model

```python
#random forest
y_pred_rfcl = rfcl.predict(x_test).ravel()
rfcl_fpr, rfcl_tpr, threshold = roc_curve(y_test, y_pred_rfcl)
auc_rfcl = auc(rfcl_fpr, rfcl_tpr)
#decision tree
y_pred_dt = dt.predict(x_test).ravel()
dt_fpr, dt_tpr, threshold = roc_curve(y_test, y_pred_dt)
auc_dt = auc(dt_fpr, dt_tpr)
#lightgbm
y_pred_lgbm = lgbm.predict(x_test).ravel()
lgbm_fpr, lgbm_tpr, threshold = roc_curve(y_test, y_pred_lgbm)
auc_lgbm = auc(lgbm_fpr, lgbm_tpr)
#Logistic Regression
y_pred_lgr = lgr.predict(x_test).ravel()
lgr_fpr, lgr_tpr, threshold = roc_curve(y_test, y_pred_lgr)
auc_lgr = auc(lgr_fpr, lgr_tpr)
#xgboost
y_pred_xgb = xgb.predict(x_test).ravel()
xgb_fpr, xgb_tpr, threshold = roc_curve(y_test, y_pred_xgb)
auc_xgb = auc(xgb_fpr, xgb_tpr)
#gradient boosting
y_pred_clf = clf.predict(x_test).ravel()
clf_fpr, clf_tpr, threshold = roc_curve(y_test, y_pred_clf)
auc_clf = auc(clf_fpr, clf_tpr)

plt.figure(figsize=(5, 5), dpi=100)
plt.plot(rfcl_fpr, rfcl_tpr, marker='.', label='RF (auc = %0.3f)' % auc_rfcl)
plt.plot(dt_fpr, dt_tpr, marker='.', label='DT (auc = %0.3f)' % auc_dt)
plt.plot(lgbm_fpr, lgbm_tpr, marker='.', label='LightGBM (auc = %0.3f)' % auc_lgbm)
plt.plot(clf_fpr, clf_tpr, marker='.', label='Gradient Boosting (auc = %0.3f)' % auc_clf)
plt.plot(lgr_fpr, lgr_tpr, marker='.', label='AdaBoost (auc = %0.3f)' % auc_lgr)
plt.plot(xgb_fpr, xgb_tpr, marker='.', label='XGBoost (auc = %0.3f)' % auc_xgb)

plt.xlabel('False Positive Rate -->')
plt.ylabel('True Positive Rate -->')

plt.legend()

plt.show()
```
