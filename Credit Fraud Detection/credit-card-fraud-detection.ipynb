### Spliting the data

```python
# split the dataset into dependent(X) and Independent(Y) datasets
x = dataset_credit_card.drop('Class', axis=1)
y = dataset_credit_card['Class']
```

```python
# Train, test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
print(x_train.values)
print(x_test.values)
print(y_train.values)
print(y_test.values)
```

```python
# Smote (Synthetic Minority Oversampling Technique)
from imblearn.over_sampling import SMOTE

print("Before oversampling: ",Counter(y_train))
SMOTE = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = SMOTE.fit_resample(x_train, y_train)
print("After oversampling:", Counter(y_train_resampled))
```

```python
# feature scaling
from sklearn.preprocessing import StandardScaler
x_train = StandardScaler().fit_transform(x_train)
x_test = StandardScaler().fit_transform(x_test)
```

```python
#classify function
def classify(model):
    model.fit(x_train, y_train)
    print("The accuracy of this model is: ", model.score(x_test, y_test)*100)
    predict = model.predict(x_test)
    print(classification_report(y_test, predict))
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
classify(lr)
```

```python
print('Logistic Regression Classifier Confusion Matrix')
y_pred = lr.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True, fmt = 'g')
```

```python
from sklearn.svm import SVC
svm = SVC()
classify(svm)
```

```python
print('SVM Classifier Confusion Matrix')
y_pred = svm.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True, fmt = 'g')
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
classify(knn)
```

```python
print('K-Nearest Neighbors Classifier Confusion Matrix')
y_pred = knn.predict(x_test)

cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm,annot=True, fmt = 'g')
```
